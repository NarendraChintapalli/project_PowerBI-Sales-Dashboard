```python
#KPI 1 query: total sales, profit, and quantity

kpi1=pd.read_sql(
  """
  select round(sum(sales),2) as total_sales,
  round(sum(profit),2) as total_profit,
  sum(quantity) as total_quantity
  from storeInfo
  """
  ,conn
)
```

```python
kpi1
```

```python
# KPI 2 Sales and Profit by Region

kpi2=pd.read_sql('''
                 select  region,round(sum(sales),2) as total_sales,
                 round(sum(profit),2) as total_profit 
                 from storeInfo group by region order by total_sales desc
                 ''',conn
                 )
```

```python
kpi2
```

```python
# KPI 3: Sales & Profit by Category & Sub-Category

kpi3=pd.read_sql(
  ''' 
  select category,[Sub-Category],round(sum(sales),2) as total_sales ,
  round(sum(profit),2) as total_profit from storeInfo group by category,[Sub-Category] order by total_sales desc
  '''
  , conn
)
```

```python
kpi3
```

```python
# Sort and take top 10
kpi3_top10 = kpi3.sort_values(by="total_sales",ascending=False).head(10)
```

```python
#plot
plt.figure(figsize=(10,6))
plt.bar(kpi3_top10["Sub-Category"],kpi3_top10["total_sales"],color="skyblue")

# Titles & labels
plt.title("Top 10 Sub-Categories by Sales", fontsize=14, fontweight="bold")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha="right")

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.bar(kpi2["Region"],kpi2["total_sales"],color="pink")

plt.title("region vs sales",fontsize=14,fontweight="bold")
plt.xlabel("region")
plt.ylabel("total_sales value",fontsize=14,fontweight="bold")

plt.xticks(rotation=45,ha="right")

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.bar(kpi2["Region"], kpi2["total_sales"], label="Total Sales", color="skyblue")
plt.bar(kpi2["Region"], kpi2["total_profit"], bottom=kpi2["total_sales"], label="Total Profit", color="orange")

plt.title("Sales + Profit by Region", fontsize=14, fontweight="bold")
plt.xlabel("Region")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python


# create x locations for regions
x = np.arange(len(kpi2["Region"]))  

plt.figure(figsize=(10,6))

# width of each bar
bar_width = 0.35  

# plot sales
plt.bar(x - bar_width/2, kpi2["total_sales"], width=bar_width, label="Total Sales", color="skyblue")

# plot profit
plt.bar(x + bar_width/2, kpi2["total_profit"], width=bar_width, label="Total Profit", color="orange")

# labels
plt.xticks(x, kpi2["Region"])
plt.title("Sales vs Profit by Region", fontsize=14, fontweight="bold")
plt.xlabel("Region")
plt.ylabel("Value")
plt.legend()
plt.show()
```

```python
#KPI-4: Top 10 Products by Sales
kpi4= pd.read_sql('''
    SELECT 
        [Product Name] AS product_name,count([product Name]) as product_freq,
        ROUND(SUM(Sales), 2) AS total_sales,
        round(sum(Profit),2) as total_profit
    FROM superstore
    GROUP BY [Product Name]
    ORDER BY total_profit DESC
    limit 10
''', conn)
kpi4

```

```python
plt.figure(figsize=(10,6))
plt.bar(kpi4["product_name"], kpi4["total_sales"], color="seagreen")
plt.title("Top 10 Products by Sales", fontsize=14, fontweight="bold")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")
plt.xticks(rotation=60,ha="right")
plt.gca().invert_yaxis()   # so top product appears at top
plt.show()
```

```python
#KPI-5: Top 10 Customers by Profit

kpi5=pd.read_sql('''
                 
                 select [Customer ID] as cust_id,[Customer Name] as cust_name,
                 round(sum(profit),2) as profits_By_Cust from storeInfo 
                 group by cust_id order by profits_By_Cust desc limit 10
                 
                 ''',conn)
```

```python
kpi5
```

```python
#KPI-6: Sales & Profit by Ship Mode

kpi6 = pd.read_sql('''
    SELECT 
        [Ship Mode] AS ship_mode,
        ROUND(SUM(sales),2) AS total_sales,
        ROUND(SUM(profit),2) AS total_profit
    FROM storeInfo
    GROUP BY [Ship Mode]
    ORDER BY total_sales DESC
''', conn)

kpi6
```

```python
#KPI-7: Monthly Sales Trend (Year + Month)

kpi7 = pd.read_sql('''
    SELECT 
        strftime('%Y', [Order Date]) AS year,
        strftime('%m', [Order Date]) AS month,
        ROUND(SUM(sales),2) AS total_sales
    FROM storeInfo
    GROUP BY year, month
    ORDER BY year, month
''', conn)

kpi7
```

```python
#KPI-8: Discount vs Profit impact

kpi8 = pd.read_sql('''
    SELECT 
        ROUND(discount,2) AS discount_level,
        ROUND(SUM(sales),2) AS total_sales,
        ROUND(SUM(profit),2) AS total_profit
    FROM storeInfo
    GROUP BY discount_level
    ORDER BY discount_level
''', conn)

kpi8
```

```python
#KPI-9: Customer Lifetime Value (CLV) - total revenue generated by the customer.

kpi9 = pd.read_sql('''
    SELECT 
        [Customer ID] AS cust_id,
        [Customer Name] AS cust_name,
        ROUND(SUM(sales),2) AS total_sales,
        ROUND(SUM(profit),2) AS total_profit,
        COUNT(DISTINCT [Order ID]) AS total_orders
    FROM storeInfo
    GROUP BY cust_id
    ORDER BY total_profit DESC
''', conn)

kpi9
```

```python
#RFM Analysis (Recency, Frequency, Monetary) KPI-10: RFM Analysis

kpi10 = pd.read_sql('''
    SELECT 
        [Customer ID]   AS cust_id,
        [Customer Name] AS cust_name,
        DATE(MAX([Order date])) AS last_purchase_date,
        [Order ID] AS order_id,
        COUNT(DISTINCT [Order ID]) AS frequency,
        ROUND(SUM(sales), 2) AS [monetary_value(total_sales)]
    FROM storeInfo
    GROUP BY cust_id, cust_name
    ORDER BY [monetary_value(total_sales)] DESC
    
''', conn)

kpi10
```
